````
# Load MNIST data
train_data = _ 
test_data  = _ 

test_data.head(5)
````

````
X_train = np.array(train_data.iloc[:, 1:])
y_train = np.array(train_data.iloc[:, 0])

X_test = np.array(test_data.iloc[:, 1:])
y_test = np.array(test_data.iloc[:, 0])

print (X_train.shape, y_train.shape, X_test.shape, y_test.shape)
````

````
# Visualize grid of samples
grid_size = (5,5)

_, axes = plt.subplots ( grid_size[0] , grid_size[1] , figsize=(12,12))
axes = axes.ravel()

for i, ax in enumerate(axes):
  ax.imshow( X_test[i].reshape(28,28) , cmap='gray')
  ax.set_axis_off(), ax.set_title("{0}".format(y_test[i]))

plt.tight_layout()
````

````
num_classes = len(np.unique(y_test))
num_inputs = X_test.reshape(10000, -1).shape[1]

print(num_classes, num_inputs)
````

````
# Reshape the data - MLPs do not understand such '2D' stuff
# Reshape to 1D is similat to "flatten" an image

# Configuration options
feature_vector_length = 784
num_classes = 10

X_train = X_train.reshape(X_train.shape[0], feature_vector_length)
X_test = X_test.reshape(X_test.shape[0], feature_vector_length)

# Convert into 0 - 1 values
X_train = X_train / 255.0
X_test = X_test / 255.0
````

````
# Convert target classes to categorical ones - one hot encoding
y_train = to_categorical(y_train, num_classes) # np.argmax(y_train, axis=1)
y_test = to_categorical(y_test, num_classes)
print(y_train.shape, y_test.shape)
````

## Create a graph model

````
# Create the model
model = Sequential()
model.add(Dense( 10 , input_shape = (784,) , activation = "softmax" ))
# model.add(Dense( 256 , activation = "relu" ))
# model.add(Dense( 128 , activation = "relu" ))
# model.add(Dense( 64, activation = "relu" ))
# model.add(Dense( 32, activation = "relu" ))
# model.add(Dense( 10, activation = "softmax"))
````

````
model.summary()
````

## Define loss function and optimizer

````
# Configure the model and start training
model.compile(loss = "categorical_crossentropy" , optimizer = "adam" , metrics = ["accuracy"] )
````

## Train model

````
# Train the model
history = model.fit( X_train , y_train , \
                    #X_test, y_test, \
                    epochs = 100 , batch_size = 200 , \
                    verbose = 1, validation_split = 0.2 )
````

## Plot results

````
# Plot progress (loss function)
_, axes = plt.subplots(1,2, figsize=(10,5))
axes[0].plot(history.history['loss'], c='b')
axes[0].plot(history.history['val_loss'], c='r')
axes[1].plot(history.history['accuracy'], c='b')
axes[1].plot(history.history['val_accuracy'], c='r')
axes[0].set_title("Loss"), axes[1].set_title('Accuracy')
````

## Compute metrics over ```X_test``` images

````
# Test the model after training
test_results = model.predict( X_test )
````

````
# test_results = np.argmax(test_results, axis=1)
y_test = np. argmax(y_test, axis = 1)
````

````
acc = accuracy_score(y_test, test_results)
cm = confusion_matrix(y_test, test_results)

print ("accuracy = {0}".format(acc))
sns.heatmap(cm, cmap='hot', annot=True)
````
