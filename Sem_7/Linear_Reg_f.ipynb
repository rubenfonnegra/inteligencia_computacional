## In Synthetic data

````python
from sklearn.datasets import make_regression
from sklearn.linear_model import LinearRegression
````

````python
X, y = make_regression(n_samples = 500, n_features = 1, noise = 10, random_state = 1)

_, ax = plt.subplots (figsize=(6,5))
ax.scatter(X, y, c='b')
print (X.shape, y.shape)
````

````python
linear_regression = LinearRegression()
linear_regression.fit(X, y)

# Imprimir los valores estimados para compararlos respecto a los originales
print ("w = " + str(linear_regression.coef_) + ", b = " + str(linear_regression.intercept_))
````

````python
# Nueva muestra. Imprime el valor correspondiente de y 
new_sample = np.array([1])

# Predecir la nueva muestra. Imprimir la prediccion
prediction = linear_regression.predict( new_sample )
print (prediction)
````

Download the [Energy efficiency](https://www.kaggle.com/datasets/elikplim/eergy-efficiency-dataset?resource=download) dataset [here](https://drive.google.com/file/d/1UpP1DoN0WQigaD0y-A75GlD-bJ2TAjMW/view?usp=drive_link) \

````python
import numpy as np
import matplotlib.pyplot as plt # plots
from sklearn.linear_model import LinearRegression # scikit-learn: libreria de IA para python
from sklearn.datasets import load_diabetes
import seaborn as sns
````

````python
# Cargar la base de datos Diabetes
diabetes_X, diabetes_y = load_diabetes(return_X_y=True)
print(diabetes_X.shape, diabetes_y.shape)
````

````python
# Use only one feature
diabetes_X = diabetes_X[:, 2:4]
# diabetes_X = diabetes_X[:, np.newaxis] 
print(diabetes_X.shape, diabetes_y.shape)
````

````python
_, axes = plt.subplots (1,2,figsize=(10,5))
sns.scatterplot(x=diabetes_X[:, 0], y=diabetes_y, ax=axes[0])
sns.scatterplot(x=diabetes_X[:, 1], y=diabetes_y, ax=axes[1])
````

````python
# Linear regression
linear_regression = LinearRegression()

# Train the model
linear_regression.fit( diabetes_X , diabetes_y )
````

````python
# Make predictions using the testing set
diabetes_y_pred = linear_regression.predict( diabetes_X )


print ("w = " + str( linear_regression.coef_ ) + ", b = " + str( linear_regression.intercept_ ))
````

````python
# Plot outputs
_, axes = plt.subplots (1,2,figsize=(10,5))
axes[0].scatter( diabetes_X[:, 0] , diabetes_y, color="blue")
axes[0].scatter( diabetes_X[:, 0] , diabetes_y_pred , color="red")

axes[1].scatter( diabetes_X[:, 1] , diabetes_y, color="blue")
axes[1].scatter( diabetes_X[:, 1] , diabetes_y_pred , color="red")
````
